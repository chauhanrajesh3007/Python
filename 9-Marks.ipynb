Problem Statement: Sports Game Analysis
In this problem, you are provided with a file named scores.txt which contains the scores of Basketball games. The file contains information about various teams and their scores for each game. The objective is to perform several analyses on the data in the file.

Your task is to:

1.	Calculate the Average Points Scored:

Compute the average of the points scored across all games listed in the file.

2.	Determine Wins and Losses for a Specific Team:

Pick your favorite team and scan through the file to determine how many games they won and how many games they lost.

3.	Identify Teams Losing by Large Margins:

Find the team(s) that lost by 30 or more points the most times.

4.	Find all Teams Losing by 70 or More Points:

Find all teams that lost by 70 or more points in any of the games.


Sample Output:

Sports Game Analysis
1. Average Points Scored
Average points scored across all games: 124.7

2. Wins and Losses for a Specific Team
Selected Team: Phoenix Suns
Games Won: 15
Games Lost: 7

3. Teams Losing by Large Margins
Teams that lost by 30 or more points the most times:
Orlando Magic: 4 times

4. Teams Losing by 70 or More Points
Teams that lost by 70 or more points in any game:
Los Angeles Lakers
New Orleans Pelicans


```python
games = []

# Read the file
with open("scores.txt", "r") as f:
    for line in f:
        parts = line.split()
        if len(parts) < 5:
            continue

        team1 = parts[1]
        score1 = int(parts[2])
        team2 = parts[3]
        score2 = int(parts[4])

        games.append((team1, score1, team2, score2))


# 1. AVERAGE POINTS
total_points = 0
total_scores = 0

for t1, s1, t2, s2 in games:
    total_points += s1 + s2
    total_scores += 2

average_points = total_points / total_scores
print("Average Points Scored:", average_points)


# 2. WINS & LOSSES FOR A TEAM
team = "NorthCarolina"   # <-- change team name here

wins = 0
losses = 0

for t1, s1, t2, s2 in games:
    if t1 == team:
        if s1 > s2:
            wins += 1
        else:
            losses += 1
    elif t2 == team:
        if s2 > s1:
            wins += 1
        else:
            losses += 1

print("\nTeam:", team)
print("Wins:", wins)
print("Losses:", losses)


# 3. TEAM LOSING BY 30+ MOST TIMES
loss_count = {}

for t1, s1, t2, s2 in games:
    diff = abs(s1 - s2)
    if diff >= 30:
        if s1 < s2:
            loser = t1
        else:
            loser = t2

        loss_count[loser] = loss_count.get(loser, 0) + 1
# Find max
max_losses = max(loss_count.values())
print("\nTeams losing by 30+ most times:")
for team, count in loss_count.items():
    if count == max_losses:
        print(team, ":", count)


# 4. TEAMS LOSING BY 70+
teams_70 = set()

for t1, s1, t2, s2 in games:
    if abs(s1 - s2) >= 70:
        if s1 < s2:
            teams_70.add(t1)
        else:
            teams_70.add(t2)

print("\nTeams losing by 70+ points:")
for t in teams_70:
    print(t)
```
