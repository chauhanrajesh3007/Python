714.BMI Calculator App
 Take user inputs:- Weight (kg) (number input)- Height (cm) (number input)
 On button click, calculate BMI = weight / (height/100)^2.- Display:
  - BMI Value
  - A health category (Underweight, Normal, Overweight, Obese)- Show results in colored messages (st.success(), st.warning(), st.error())
  Underweight (below 18.5), Normal (18.5–24.9), Overweight (25.0–29.9), and Obese (30.0 or higher
  

```python
%%writefile pb714.py
import streamlit as st

weight=st.number_input("Enter a weight:")
height=st.number_input("Enter a height:")

if st.button("BMI"):
    BMI=weight/((height/100)**2)
    st.write("BMI Value:",BMI)
    if BMI<18.5:
        st.warning("Underweight")
    elif 18.5<=BMI<=24.9:
        st.success("Normal")
    elif 25<=BMI<=29.9:
        st.warning("Overweight")
    elif BMI>=30:
        st.error("Obese")
```

```python
%%writefile ex.py
import streamlit as st
from datetime import date
import pandas as pd

st.set_page_config(
page_title="Hello Streamlit",
layout="centered"
)

st.subheader("Information")

col1,col2=st.columns([1,2])#1:2 ratio

with col1:
    enrollment_no=st.text_input("Enter a Enrollment Number:","24002171610003")
    name=st.text_input("Enter a name:","Rajesh")
with col2:
    semester=st.selectbox("semester",["3","4","5","6","7","8"])
    div=st.text_input("Enter a division:","CSE")
    
exam_date=st.date_input("Select Exam Date:",value=date.today())

st.subheader("Marks Entry")

p=st.text_input("Enter a Python Marks:")
f=st.text_input("Enter a FSD Marks:")
d=st.text_input("Enter a DE Marks:")

st.subheader("Feedback")
understand=st.slider("Understanding:",0,10,5)
participation=st.radio("Class Participation",["Low","Medium","High"])
comments=st.text_area("Enter Your Feedback or Comment")

if st.button("submit record"):
    if not enrollment_no or not name:
        st.error("pls fill enrollment number and student name.")
    else:
        df=pd.DataFrame({"Enrollment":[enrollment_no],
                        "Name":[name],
                        "Sem":[semester],
                        "Division":[div],
                        "Exam-Date":[exam_date],
                        "Python-1":[p],
                        "FSD-1":[f],
                        "DE":[d],
                        "Ubderstanding":[understand],
                        "Participation":[participation],
                        "Comments":[comments]})
    st.success("Record submitted successfully..!!")
    st.subheader("Preview")
    st.dataframe(df)
    csv=df.to_csv(index=False).encode("utf-8")

    st.download_button(
    label="Download as CSV",
    data=csv,
    file_name="sample_marks.csv",
    mime="text/csv")
```
